```python
def load_data(data_type, participant, trial):
    filename = "%s_%s_%s.csv" % (data_type, participant, trial)
    return pd.read_csv("../extracted_motion_parameters/data/" + filename, index_col=0, header=None, squeeze=True)

from collections import defaultdict
velocity = defaultdict(dict)
for participant in analyze_mocap.participants:
    for trial in analyze_mocap.trials:
        velocity[participant][trial] = load_data("velocity", participant, trial)
```

```python
# first_frame = analyze_mocap.get_frame_range("Matan", 1).start
normalized = [
    analyze_mocap.normalize_frame_numbers(velocity, participant, trial)
    for participant in analyze_mocap.participants
    for trial in analyze_mocap.trials
]
velocity_normalized = pd.concat(normalized, axis='columns')
velocity_normalized.columns = pd.MultiIndex.from_product([analyze_mocap.participants, analyze_mocap.trials], names=["Participant", "Trial"])
#frame_count = 4 * 8 * 4 * 100
# last_frame = first_frame + frame_count
# velocity["Matan"][1].iloc[first_frame:last_frame].reindex(range(1212))
#velocity_normalized = pd.DataFrame(, index=np.arange(frame_count), columns=pd.MultiIndex.from_product([analyze_mocap.participants, analyze_mocap.trials]))
```
